```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
#select player according to skill move
player_data = pd.read_csv('fifa_data.csv',index_col=0)
most_skill_player = player_data[player_data['Skill Moves'] == 5]
most_skill_player.to_csv('most_skill_player.csv', index = False)
```

```python
most_skill_player = pd.read_csv('most_skill_player.csv')
#drop columns
most_skill_player.drop(['Photo','Flag','Club Logo'], axis = 1, inplace = True)
most_skill_player
```

```python
#scatter of most potenstial players
fig, ax = plt.subplots(figsize = (10,5), layout = 'constrained')
ax.scatter(most_skill_player.Age,most_skill_player.Potential,s = 10, color = "red")
ax.set_xlabel('age')
ax.set_ylabel('Potential')
ax.set_title('age and potentiality of most skill players')
```

```python
clubs = pd.DataFrame(most_skill_player['Club'])
#counts values and map accordint to data
num_club = clubs['Club'].value_counts()
clubs['n_club'] = clubs['Club'].map(num_club)
#drop duplicates
final_club = clubs.drop_duplicates()
#drop null values
final_clubs = final_club.dropna()
#clubs.drop_duplicates()
```

```python
#reset index
final_clubs.reset_index(drop=True)
club_num = final_clubs.set_index(np.arange(41))
club_num
#club_num['n_club'] = club_num['n_club'].astype(int)
```

```python
#draw bar accordint to club players
fig, bx = plt.subplots(figsize=(50, 10), layout='constrained')
bx.bar(club_num['Club'],club_num['n_club'])
bx.set_xlabel('name of clubs')
```

```python
overall_skill = player_data[player_data['Overall'] >= 85]
overall_skill.to_csv('overall_skill_player.csv', index =False)
overall_skill_player = pd.read_csv('overall_skill_player.csv')
overall_skill_player.drop(['Photo','Flag','Club Logo'] ,axis = 1, inplace = True)
overall_skill_player
```

```python
overall = pd.DataFrame(overall_skill_player['Nationality'])
le = LabelEncoder() 
overall_counts = overall['Nationality'].value_counts()
overall['n_count'] = overall['Nationality'].map(overall_counts)
overall_final = overall.drop_duplicates().dropna()
overall_final 
```

```python
overall_f = overall_final[overall_final['n_count'] > 3]
fig, px = plt.subplots(figsize = (10,10))
plt.pie(overall_f['n_count'], labels = overall_f['Nationality'], shadow = True)
plt.title('overall ranking players according to nationality')
plt.legend(title = 'Country list', loc = 'upper right')
```
